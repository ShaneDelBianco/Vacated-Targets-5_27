```python
full_df = full_df.drop(280)
full_df = full_df.drop(283)

```

```python
import pandas as pd
full_df["total_air_yards"] = full_df["receiver_air_yards"] * full_df["targets"]

pd.set_option('chained_assignment', None)

lag_features = ["total_air_yards"]

for lag in range(1, 4):

    shifted = full_df.groupby('player_id').shift(lag)
    
    for column in lag_features:
        
        full_df[f'lag_{column}_{lag}'] = shifted[column]
        
full_df = full_df.fillna(0)
```

```python
"""Get rid of rookies"""
vacated_df = full_df.loc[(full_df["lag_Team_1"] != -1)]

"""No players who stayed on the same team"""
vacated_df = vacated_df.loc[(vacated_df["lag_Team_1"] != vacated_df["Team"])]

"""Corrects season for hold outs and season long injuries"""
vacated_df["season"] = vacated_df["season"].apply(int)
vacated_df["lag_season_1"] = vacated_df["lag_season_1"].apply(int)

for index, row in vacated_df.iterrows():
    if row['season'] != row["lag_season_1"] + 1:
        vacated_df["season"] = vacated_df["lag_season_1"] +1
        
#vacated_df = vacated_df.loc[(vacated_df["season"] == vacated_df["lag_season_1"]+1)]

"""adding retirement year players. Will screw up 2020 stats but those are already messed up"""
retired_df = full_df.loc[(full_df["season"] == full_df["retirement_year_correct"])]

"""For players that play 1 year and retire"""
for index, row in retired_df.iterrows():
    if row["lag_Team_1"] == -1:
  #      row["lag_Team_1"] = row["Team"]
        retired_df['lag_Team_1'] = retired_df["Team"]
"""The problem is that for retired players, we don't want their lag1 targets. We want their final year stats added to the teams
next season totals."""


"""People who retired for a year or 2 and then returned to the same team"""
retire_return = full_df.loc[full_df["lag_season_1"] != full_df["season"] - 1]
retire_return = retire_return.loc[retire_return["Team"] == retire_return["lag_Team_1"]]
jason_witten = retire_return.loc[retire_return["player_name_x"] == "J.Witten"]
jason_witten
```
